```python
from itertools import permutations #kode disamping digunakan untuk membuat kombinasi barang
rules = list(permutations(df.columns, 2)) #kode disamping digunakan untuk menyimpan seluruh kemungkinan kombinasi barang
rules #kode disamping digunakan untuk menampilkan seluruh hasil dari kombinasi barang
```

```python
rule_baru = [] #kode disamping digunakan untuk menyimpan data dalam bentuk list
for rule in rules: #kode disamping digunakan untuk mengulangi pembacaan dataset
  if len(df[(df[rule[0]]==1) & (df[rule[1]]==1)]) >0: #kode disamping digunakan untuk menentukan kombinasi tersebut ada atau tidak pada data
    rule_baru.append(rule) #kode disamping digunakan untuk menyimpan hasil pengecekan dalam bentuk tuple di dalam list
rule_baru #kode disamping digunakan untuk menampilkan hasil kombinasi
```

```python
hitung = {} #kode disamping digunakan untuk menyimpan data dalam bentuk dictionary
for rule in rule_baru: #kode disamping digunakan untuk mengulangi pembacaan dataset
  banyak_data= len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1) #kode disamping digunakan untuk menentukan jumlah kemunculan setiap kombinasi lalu menyimpan datanya ke dalam sebuah variabel
      ]) 
  hitung[str(rule)] = banyak_data #kode disamping digunakan untuk menambahkan hasil penentuan ke dalam dictionary
hitung #kode disamping digunakan untuk menampilkan hasil kombinasi
```

##Support dan Confident

```python
for item,ab in hitung.items():
  print(f"{item} support = {ab/len(df)*100}%")
```

```python
for item,ab in hitung.items():
  i = eval(item)
  print(f"{i[0]}->{i[1]}  Confidence = {ab/df[i[0]].sum()*100}%")
```

##1. Tambahkan kolom jarak untuk menampilkan jarak dari dataset terhadap target prediksi

```python
# membuat tabel jarak

import numpy as np

# kode dibawah digunakan untuk membuat dataframe dari data yang baru
new_data = pd.DataFrame({'diagonal'    : [172.25],
                         'height_left' : [103.94],
                         'height_right': [103.68],
                         'margin_low'  : [4.14],
                         'margin_up'   : [3.25],
                         'length'      : [112.48]})

# kode dibawah digunakan untuk menghitung jarak antar setiap baris dataframe utama dengan dataframe data baru
distances = []
for i in range(len(df2)):
    row = df2.iloc[i]
    distance = np.sqrt((row['diagonal'] - new_data['diagonal']) ** 2 +
                       (row['height_left'] - new_data['height_left']) ** 2 +
                       (row['height_right'] - new_data['height_right']) ** 2 +
                       (row['margin_low'] - new_data['margin_low']) ** 2 +
                       (row['margin_up'] - new_data['margin_up']) ** 2 +
                       (row['length'] - new_data['length']) ** 2)
    distances.append(distance)

# kode dibawah digunakan untuk menambahkan kolom jarak ke dataframe utama
df2['distance'] = distances

# kode dibawah digunakan untuk menampilkan tabel jarak
print('Tabel Jarak : ')
print(df2['distance'])
```

##2. Urutkan data berdasarkan K jarak terdekat (Untuk K nya ambil akar 2 dari banyak data)

```python
# kode dibawah digunakan untuk menghitung jarak antar setiap baris dataframe utama dengan dataframe data baru
df2['distance'] = np.sqrt(
    (df2['diagonal'] - 172.25)**2
    + (df2['height_left'] - 103.94)**2
    + (df2['height_right'] - 103.68)**2
    + (df2['margin_low'] - 4.14)**2
    + (df2['margin_up'] - 3.25)**2
    + (df2['length'] - 112.48)**2
)

# kode dibawah digunakan untuk mengurutkan data dari kecil ke besar dan menentukan data terdekat
k = int(np.sqrt(len(df2)))
data_terdekat = df2.sort_values('distance').head(k)

# kode dibawah digunakan untuk menampilkan data terdekat
print(data_terdekat)
```
